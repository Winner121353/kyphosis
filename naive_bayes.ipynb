```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import numpy as np
```

```python
data = pd.read_csv('bank.csv')
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb_x1 = LabelEncoder()
x1 = lb_x1.fit_transform(data.job)
lb_x2 = LabelEncoder()
x2 = lb_x2.fit_transform(data.marital)
lb_x3 = LabelEncoder()
x3 = lb_x3.fit_transform(data.education)
lb_x4 = LabelEncoder()
x4 = lb_x4.fit_transform(data.housing)
lb_x5 = LabelEncoder()
x5 = lb_x5.fit_transform(data.loan)
```

```python
data.job = x1
data.marital = x2
data.education = x3
data.housing = x4
data.loan = x5
data.head()
```

```python
data.job.value_counts()
```

```python
X = data.drop(['loan'],axis=1)
y = data['loan']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
m = GaussianNB()
```

```python
m.fit(X_train,y_train)
```

```python
pred = m.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(pred,y_test))
```
